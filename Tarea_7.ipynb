```python
# Problema 3.3
SERIE= 0
N= int(input("Ingrese un número entero: "))
BAND="T"
I= 1
while I<=N:
    if BAND=="T":
        SERIE= SERIE +(1/I)
        BAND= "F"
    else:
        SERIE= SERIE -(1/I)
        BAND="T"
    I=I+1
print(SERIE)
```
